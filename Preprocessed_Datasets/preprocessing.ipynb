```python
# providers_data.csv cleaning
#load the Raw_Datasets


providers_df = pd.read_csv(r'D:\PROJECTS\Local Food Wastage Management System\Raw_Datasets\providers_data.csv')
df_3 = providers_df.copy()
df_3
#Getting basic information about the dataset
print("ðŸ“ŒFirst 5 rows : \n", df_3.head())
print("ðŸ“ŒStructure : \n", df_3.shape)
print("ðŸ“ŒInformation : ")
print(df_3.info())
print("ðŸ“ŒNull values rows count : \n", df_3.isnull().sum())
print("ðŸ“ŒDuplicate rows count : /n", df_3.duplicated().sum())
#Cleaning Phone numbers
def clean_phone(phone):
    digits = re.sub(r'\D', '', str(phone))
    if len(digits) >= 10:
        return f"{digits[-10:-7]}-{digits[-7:-4]}-{digits[-4:]}"
    return None
df_3['Contact'] = df_3['Contact'].apply(clean_phone)
df_3['Contact']
df_3.to_csv('./cleaned_providers_data.csv', index=False)
```
